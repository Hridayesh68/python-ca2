```python
import pandas as pd
data=pd.read_csv('water_pollution_disease.csv')
data
```

```python
Problem Statement: Water pollution has been a growing concern for public health. This dataset aims to analyze the relationship between water pollution levels and the incidence of diseases across various regions. The goal is to:

Identify patterns and correlations between polluted water sources and disease outbreaks.

Visualize how different pollutants impact public health.

Clean and preprocess the data to ensure accuracy and usability for further analysis or predictive modeling.
```

```python
print(" Data Cleaning  and Data Analysis")
print("Shape of the dataset:", data.shape)
print("\nColumn names:\n", data.columns)
print("\nSample data:\n", data.head())
print("\nMissing values:\n", data.isnull().sum())
print("\nData types:\n", data.dtypes)
```

```python
Initial Data Analysis Summary
The dataset contains:

Shape: 3000 rows × 24 columns

No missing values ✅

Data types: A mix of object, int64, and float64 types. All appear correctly typed.
```

```python

duplicates = data.duplicated().sum()
data_cleaned = data.drop_duplicates()


categorical_cols = data_cleaned.select_dtypes(include='object').columns
unique_values = {col: data_cleaned[col].unique() for col in categorical_cols}


describe = data_cleaned.describe()

duplicates, unique_values, describe
```

```python
Data Cleaning Summary
No duplicate rows found.

Categorical values are consistent:

Countries: 10 unique (e.g., Mexico, Brazil, India, etc.)

Regions: North, South, East, West, Central

Water Source Types: Lake, Well, Pond, Tap, River, Spring

Water Treatment Methods: Filtration, Boiling, None, Chlorination

All values appear within logical ranges. We'll verify statistical outliers in visualizations.
```

```python
Visualization & EDA


Distribution of water contaminants (e.g., pH, turbidity, lead)

Disease incidence vs. water quality

Correlation heatmap between water quality and health indicators

Country-wise comparisons.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


sns.set(style="whitegrid")
plt.figure(figsize=(16, 10))


fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.histplot(data_cleaned['pH Level'], bins=30, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution of pH Level')

sns.histplot(data_cleaned['Turbidity (NTU)'], bins=30, kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Distribution of Turbidity (NTU)')

sns.histplot(data_cleaned['Lead Concentration (µg/L)'], bins=30, kde=True, ax=axes[1, 0], color='green')
axes[1, 0].set_title('Distribution of Lead Concentration')

sns.histplot(data_cleaned['Bacteria Count (CFU/mL)'], bins=30, kde=True, ax=axes[1, 1], color='orange')
axes[1, 1].set_title('Distribution of Bacteria Count')

plt.tight_layout()
plt.show()
```

```python

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.scatterplot(data=data_cleaned, x='Lead Concentration (µg/L)', y='Diarrheal Cases per 100,000 people', ax=axes[0], color='red')
axes[0].set_title('Lead vs Diarrheal Cases')

sns.scatterplot(data=data_cleaned, x='Bacteria Count (CFU/mL)', y='Cholera Cases per 100,000 people', ax=axes[1], color='purple')
axes[1].set_title('Bacteria Count vs Cholera Cases')

sns.scatterplot(data=data_cleaned, x='Turbidity (NTU)', y='Typhoid Cases per 100,000 people', ax=axes[2], color='blue')
axes[2].set_title('Turbidity vs Typhoid Cases')

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(16, 12))
correlation_matrix = data_cleaned.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap: Water Quality, Health & Socioeconomic Indicators")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")


plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x="Water Source Type", y="Cholera Cases per 100,000 people",hue="Water Source Type",palette="Set2")


plt.title("Cholera Cases per 100,000 People by Water Source Type", fontsize=14)
plt.xlabel("Water Source Type", fontsize=12)
plt.ylabel("Cholera Cases per 100,000 People", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python



water_source_counts = data["Water Source Type"].value_counts()


cholera_by_year = data.groupby("Year")["Cholera Cases per 100,000 people"].mean()


avg_contaminant_by_region = data.groupby("Region")["Contaminant Level (ppm)"].mean().sort_values()


fig, axs = plt.subplots(1, 3, figsize=(20, 6))


axs[0].pie(water_source_counts, labels=water_source_counts.index, autopct='%1.1f%%', startangle=140)
axs[0].set_title("Distribution of Water Source Types")


axs[1].plot(cholera_by_year.index, cholera_by_year.values, marker='o', linestyle='-', color='teal')
axs[1].set_title("Average Cholera Cases Over Years")
axs[1].set_xlabel("Year")
axs[1].set_ylabel("Cholera Cases per 100,000 People")


axs[2].bar(avg_contaminant_by_region.index, avg_contaminant_by_region.values, color='coral')
axs[2].set_title("Average Contaminant Level by Region")
axs[2].set_xlabel("Region")
axs[2].set_ylabel("Contaminant Level (ppm)")
axs[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
